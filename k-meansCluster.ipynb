```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import scale
from sklearn import datasets
import pandas as pd

import sklearn.metrics
from sklearn.metrics import confusion_matrix, classification_report
```

```python
iris=datasets.load_iris()
X=scale(iris.data) # Standardize the dataset
y=pd.DataFrame(iris.target)
varaible_name=iris.feature_names

table=pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
clustering=KMeans(n_clusters=3, random_state=5)
clustering.fit(X)
```

```python
color_1=np.array(['red','blue','green'])


plt.subplot(1,2,1)
plt.scatter(x=table['petal length (cm)'], y=table['petal width (cm)'], c=color_1[iris.target]) 
plt.title('Actual')

plt.subplot(1,2,2)
plt.scatter(x=table['petal length (cm)'], y=table['petal width (cm)'], c=color_1[clustering.labels_])
plt.title('Prediction')
plt.show()
```

```python
axes1=np.choose(clustering.labels_, [2,0,1]) # changed the labels in the array
# 0-->2
# 1-->0
# 2-->1
```

```python
# Plot with corect labels
color_1=np.array(['red','blue','green'])
plt.subplot(1,2,1)
plt.scatter(x=table['petal length (cm)'], y=table['petal width (cm)'], c=color_1[iris.target]) 
plt.title('Actual')

plt.subplot(1,2,2)
plt.scatter(x=table['petal length (cm)'], y=table['petal width (cm)'], c=color_1[axes1])
plt.title('Prediction')
plt.show()
```

```python
tabel=classification_report(y,axes1)
print(tabel)
```
