```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import datasets
from sklearn.metrics import classification_report
```

```python
data_set=pd.read_csv('heart.csv')

print(data_set.isnull().sum())
data_set
```

```python
x=data_set.drop('target', axis=1)
y=data_set['target']

print(f"Shape of x: {x.shape}")
print(f"Shape of y: {y.shape}\n")
x_train,x_test,y_train,y_test=train_test_split(x,y, train_size=0.8, random_state=42)

print(f"Shape of x-train: {x_train.shape}")
print(f"Shape of y-train: {y_train.shape}")
print(f"Shape of x-test: {x_test.shape}")
print(f"Shape of y-test: {y_test.shape}")
```

```python
model=RandomForestClassifier(n_estimators=10, criterion='entropy') # n_estimators equivalent to no. of decision tress to deploy for the classification
model.fit(x_train,y_train)
```

```python
predict_result = model.predict(x_test)
print(f"Shape of prediction: {predict_result.shape}")
print(f"Actual shape: {y_test.shape}")
```

```python
data_dict = {
    'Actual': y_test,
    'Predicted': predict_result
}
pd.DataFrame(data_dict)
```

```python
print(f"Train accuracy: {model.score(x_train,y_train)}")
print(f"Test accuracy: {model.score(x_test,y_test):.3f}")
```

```python
print(classification_report(predict_result,y_test))
```
